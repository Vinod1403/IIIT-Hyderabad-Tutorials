**Step 1: Loading and Preparing the Data**

First, we'll import the necessary libraries and load the dataset from a URL.

```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

**Step 2: Data Cleaning**

```python
data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))
```

```python
X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

We can see that a number of rows with missing data were removed.

**Step 3: Separating Features and Target**


```python
plt.figure(figsize=(20, 6))
plt.scatter(X["make"], y, color='orange', marker='*', s=150, alpha=0.7, edgecolors='black')
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.6)
plt.title('Car Maker vs Price (Customized Scatter Plot)', fontsize=15, fontweight='bold')
plt.xlabel('Car Brand')
plt.ylabel('Price (in $)')
plt.show()
```

## BOX PLOT

```python
sns.set(rc={'figure.figsize':(24,7)})
sns.boxplot(x=X["make"],y=y, palette="Set3").set_title('Car Manufacturer vs Price - Box Plot')
```

```python
sns.set(style="whitegrid", rc={'figure.figsize':(20,7)})
sns.boxplot(x=X["make"], y=y, palette="coolwarm", width=0.6, linewidth=2)
plt.xticks(rotation=90)
plt.title('Car Manufacturer vs Price (Enhanced Box Plot)', fontsize=14, fontweight='bold')
plt.xlabel('Car Brand')
plt.ylabel('Price')
plt.show()
```

### Interpretation of the Box Plot

- **High-End Brands:**
  It's now crystal clear that **Mercedes-Benz, Jaguar, and Porsche** have the highest median prices. Their entire "boxes" are higher than most other brands, indicating consistently premium pricing.

- **Price Range:**
  Look at **BMW**â€”the box is very tall, meaning its prices are widely spread out. This suggests BMW offers a broad range of models, from more affordable to very expensive.

- **Outliers:**
  We can spot outliers easily. For example, there's a **Mazda, a Mitsubishi, and a Toyota** model priced significantly higher than the rest of their brand's lineup. This could be a high-performance sports model or a luxury vehicle.

**Conclusion:**
This single box plot gave us far more insight than the scatter plot, highlighting both central tendencies and variability within each brand.


**The Violin Plot: Box Plot + Density**




```python
sns.violinplot(x=X["make"],y=y, palette="Set3").set_title('Car maker vs Price - Violin Plot')
```

```python
sns.set(style="darkgrid", rc={'figure.figsize':(20,7)})
sns.violinplot(x=X["make"], y=y, palette="magma", inner="quartile", scale="count", bw=0.3)
plt.xticks(rotation=90)
plt.title('Car Maker vs Price (Violin Plot with KDE and Quartiles)', fontsize=14)
plt.show()
```

### Interpretation of the Violin Plot

- **Jaguar:**
  The violin is fairly narrow and centered high up, indicating that Mercedes' cars are **consistently expensive** with little price variation.

- **Nissan:**
  The violin is wide at the bottom and has a long tail, showing that most of Nissan's cars are **low-priced**, but they also have a few more expensive models.

**Conclusion:**
This plot gives us a much more **nuanced view** of the price distribution for each brand, revealing both concentration and spread of prices.


Another plot type are the **Swarm Plots** (Every Point Matters)


```python
sns.swarmplot(x=X["make"],y=y).set_title('Car maker vs Price Swarm Plot')
```

```python
sns.set(rc={'figure.figsize':(22,7)})
sns.swarmplot(x=X["make"], y=y, hue=X["fuel-type"], palette="husl", dodge=True, size=5)
plt.xticks(rotation=90)
plt.title('Car Maker vs Price by Fuel Type (Swarm Plot)', fontsize=14, fontweight='bold')
plt.show()
```

**Note:** This plot can be slow and hard to read for very large datasets, but for our cleaned dataset of 159 cars, it works well.

Now let us try to learn a bit more about the various features in our dataset. One of the features is **Horsepower**, used to describe the power that an engine is capable of producing. A sensible assumption could be that Sport car companies like Porsche would have more cars with higher horsepowers. Lets check if this is actually true ...

```python
sns.boxplot(x=X["make"],y=pd.to_numeric(X["horsepower"]), palette="Set3").set_title('Car maker vs Horsepower Box Plot')
```

Interpretation:

As expected, Porsche and Jaguar lead the pack with very high horsepower cars. This confirms our intuition. Now, the next logical question is: does horsepower have a direct relationship with our target variable, **price**?

A simple way to check this is to again plot a scatterplot (but this time lets do this using Seaborn instead of Matplotlib).

An expected positive correlation should show up with most of the points lying around a line with a positive slope.

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.scatterplot(x=pd.to_numeric(X["horsepower"]), y = y, color = "red")
```

A better way of plotting these correlations is using a **Jointplot** instead of a simple scatter plot. The jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins.

I have also used the parameter "reg" which fits a regression line through the points making the correlation sign apparent. A line with positive slope would indicate a postiive correlation and vice versa. A line parallel to x-axis would indicate no correlation between the 2 variables (implying independence).

```python
sns.jointplot(x=pd.to_numeric(X["horsepower"]), y = y, kind="reg", color = 'green')
```

```python
sns.jointplot(x=pd.to_numeric(X["horsepower"]), y=y, kind="hex", color="coral", height=7, space=0.3)
plt.suptitle('Horsepower vs Price (Hexbin Joint Plot)', fontsize=14, y=1.02)
```

```python
import requests
from io import StringIO
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
txt = requests.get(url).text
df = pd.read_csv(StringIO(txt))
df.replace("?", np.nan, inplace=True)
df = df.dropna()
numeric_cols = ['symboling', 'wheel-base', 'length', 'width', 'height', 'curb-weight',
                'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower',
                'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
for c in numeric_cols:
    df[c] = pd.to_numeric(df[c])
corr_with_price = df[numeric_cols].corr()['price'].sort_values()
print("Numeric features sorted by correlation with price (most negative first):\n")
print(corr_with_price)
```

```python
feature = "highway-mpg"
sns.set(style="whitegrid")
g = sns.jointplot(x=df[feature], y=df['price'], kind="reg", height=7, scatter_kws={'alpha':0.5, 's':50})
g.ax_joint.set_title(f'{feature} vs Price (Jointplot with Regression)', fontsize=14, pad=20)
plt.show()
```

```python
# hexbin (good when points overlap)
sns.set()
g = sns.jointplot(x=df[feature], y=df['price'], kind="hex", height=7, marginal_ticks=True)
g.ax_joint.set_xlabel(feature)
g.ax_joint.set_ylabel("price")
g.ax_joint.set_title(f'{feature} vs Price (Hexbin Joint Plot)', fontsize=14, pad=20)
plt.show()
```

```python
import scipy.stats as stats
sns.set(style="darkgrid")
g = sns.jointplot(x=df[feature], y=df['price'], kind="reg", height=7, scatter_kws={'s':40, 'alpha':0.6}, line_kws={'color':'red'})
r, p = stats.pearsonr(df[feature].dropna(), df['price'].dropna())
ax = g.ax_joint
ax.text(0.02, 0.95, f'Pearson r = {r:.3f}\np-value = {p:.3e}', transform=ax.transAxes,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
ax.set_title(f'{feature} vs Price (Regression with Pearson r)', fontsize=14, pad=20)
plt.show()
```

## Short interpretation
Interpretation: The jointplot for highway-mpg vs price shows a clear negative correlation (Pearson r â‰ˆ <value from your run>). The regression line slopes downward indicating that vehicles with higher fuel efficiency (higher mpg) tend to have lower prices. This is sensible: performance / luxury cars â€” which are more expensive â€” usually trade off fuel economy for power, while economy cars (lower price) have higher mpg. The hexbin plot (or KDE marginals) further shows concentration of low-price/high-mpg vehicles.

## Question: Interpreting Body Style vs. Price
Let's analyze the plot below, which shows the relationship between a car's body-style and its price.

Based on the plot, answer the following questions:

- Which body-type has the highest median price? Which one appears to be the "costliest" overall?
- Is looking only at the maximum value (the top whisker or outlier) a good way to determine the costliest body-style? Why is the interquartile range (the box) a better indicator?
- Do you think the body-style of a car could be a good predictor for its price? Why or why not?


```python
sns.boxplot(x=X["body-style"],y=y, palette="Set3").set_title('Body style vs Price Box Plot')
```



1. Which body-type has the highest median price? Which one appears to be the costliest overall?

From the Box Plot of Body Style vs. Price,

The convertible body style shows the highest median price, as its box (the central 50% of data) is located higher on the price axis compared to others.

Overall, convertibles and hardtops appear to be the costliest body styles, indicating that these luxury-oriented car types generally have higher price ranges.

This aligns with real-world expectations since convertibles and hardtops are often premium models with more expensive design and materials.

2. Is looking only at the maximum value (the top whisker or outlier) a good way to determine the costliest body-style? Why is the interquartile range (the box) a better indicator?
Ans:No, looking only at the maximum value or an outlier is not a reliable way to determine the costliest body-style.

The top whisker or an outlier might represent a single rare car that is unusually expensive (for example, a sports or limited-edition model).

It does not represent the general price trend for that body-style.

The interquartile range (IQR) â€” the box portion of the plot â€” gives a better indication of the typical price range for that body type because:

It covers the middle 50% of prices, excluding extreme highs and lows.

It helps compare which body styles are consistently more expensive, rather than being affected by single exceptional cases.

So, a body-style with a higher median and higher IQR box generally indicates a more expensive category overall.



3. Do you think the body-style of a car could be a good predictor for its price? Why or why not?

The body-style can be a useful but not a strong predictor of a carâ€™s price.

It provides some insight since certain styles (like convertibles or sedans) are typically designed for specific market segments â€” luxury vs. economy â€” thus influencing the price.

However, price is also affected by many other factors, such as brand, engine size, horsepower, features, and fuel type.

Two cars with the same body-style can vary widely in price depending on these attributes.

Hence, body-style alone gives a rough trend but not an accurate prediction. It should be combined with other features (like horsepower, engine-size, and brand) in predictive models for better results.
