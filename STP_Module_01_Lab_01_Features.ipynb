```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
params = [
    {"n_neighbors": 3, "weights": "uniform", "metric": "euclidean"},
    {"n_neighbors": 5, "weights": "distance", "metric": "euclidean"},
    {"n_neighbors": 7, "weights": "uniform", "metric": "manhattan"},
    {"n_neighbors": 9, "weights": "distance", "metric": "minkowski"},
]
for p in params:
    classifier = KNeighborsClassifier(
        n_neighbors=p["n_neighbors"], weights=p["weights"], metric=p["metric"]
    )
    classifier.fit(X_train, y_train)
    predictions = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    print(f"--- KNN Parameters ---")
    print(f"n_neighbors: {p['n_neighbors']}, weights: {p['weights']}, metric: {p['metric']}")
    print(f"Classification Accuracy: {accuracy:.2%}")
    print(f"Predicted classes: {predictions[:5]}")
    print(f"Actual classes: {y_test[:5]}")
    print("-" * 60)
```

### Why CSV Format?

CSV (Comma-Separated Values) is one of the most common formats for ML datasets because:

1. **Structured**: Data is organized in rows and columns (tabular format)
2. **Easy to read**: Both humans and machines can parse it easily
3. **Universal**: Works with all ML libraries (scikit-learn, pandas, TensorFlow, etc.)
Each row = one sample/observation
Each column = one feature
Last column = target/label (what we want to predict)

### Converting Raw Data to Features

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
import numpy as np
texts = ["I love this!", "Terrible product", "Amazing quality"]
# Bag-of-Words (simple count)
vectorizer_bow = CountVectorizer()
features_bow = vectorizer_bow.fit_transform(texts)
df_bow = pd.DataFrame(features_bow.toarray(), columns=vectorizer_bow.get_feature_names_out())
print("=== Bag-of-Words ===")
print(df_bow)
# TF-IDF Features
vectorizer_tfidf = TfidfVectorizer()
features_tfidf = vectorizer_tfidf.fit_transform(texts)
df_tfidf = pd.DataFrame(features_tfidf.toarray(), columns=vectorizer_tfidf.get_feature_names_out())
print("\n=== TF-IDF Features ===")
print(df_tfidf)
#Character-level N-grams (captures subword info)
vectorizer_char = CountVectorizer(analyzer='char', ngram_range=(2,3))
features_char = vectorizer_char.fit_transform(texts)
df_char = pd.DataFrame(features_char.toarray(), columns=vectorizer_char.get_feature_names_out())
print("\n=== Character-level N-grams ===")
print(df_char)
#Keyword One-Hot Encoding (custom important words)
keywords = ['love', 'terrible', 'amazing', 'good', 'bad']
# Generate one-hot manually since it's domain-specific
features_keywords = []
for text in texts:
    features_keywords.append([1 if k in text.lower() else 0 for k in keywords])
df_keywords = pd.DataFrame(features_keywords, columns=keywords)
print("\n=== Keyword One-Hot Encoding ===")
print(df_keywords)
#Text length / word count as numeric features
df_length = pd.DataFrame({
    'char_count': [len(t) for t in texts],
    'word_count': [len(t.split()) for t in texts]
})
print("\n=== Text Length / Word Count Features ===")
print(df_length)
#Combine all features into a single DataFrame
df_combined = pd.concat([df_bow, df_tfidf, df_char, df_keywords, df_length], axis=1)
print("\n=== Combined Features ===")
print(df_combined)
```

```python
!pip install wikipedia -q
import wikipedia
topic1 = 'Giraffe'
topic2 = 'Elephant'
def get_wikipedia_content(topic, lang='en', auto=False):
    """
    Safely fetch Wikipedia content with error handling.
    Includes auto-suggest toggle and clearer status messages.
    """
    wikipedia.set_lang(lang)
    try:
        page = wikipedia.page(topic, auto_suggest=auto)
        print(f"‚úì Fetched '{page.title}' [{lang}] successfully")
        return page.content
    except wikipedia.exceptions.DisambiguationError as e:
        print(f"‚ö†Ô∏è Disambiguation for '{topic}' in {lang}: using '{e.options[0]}'")
        return wikipedia.page(e.options[0]).content
    except wikipedia.exceptions.PageError:
        if not auto:
            print(f"‚ùå Page not found for '{topic}' in {lang}, retrying with auto_suggest=True...")
            return get_wikipedia_content(topic, lang, auto=True)
        else:
            print(f"üö´ No valid page found for '{topic}' in {lang}")
            return ""
    except Exception as e:
        print(f"‚ö†Ô∏è Unexpected error for '{topic}' [{lang}]: {e}")
        return ""
eng1 = get_wikipedia_content(topic1, 'en')
eng2 = get_wikipedia_content(topic2, 'en')
fr1 = get_wikipedia_content(topic1, 'fr')
fr2 = get_wikipedia_content(topic2, 'fr')
print("\n‚úÖ All Wikipedia pages fetched successfully!")
print(f"üìò English {topic1} length: {len(eng1)} characters")
print(f"üìó English {topic2} length: {len(eng2)} characters")
print(f"üìô French  {topic1} length: {len(fr1)} characters")
print(f"üìí French  {topic2} length: {len(fr2)} characters")

# Optional preview (first 300 characters)
print("\n--- Sample Preview ---")
print(f"\nüîπ {topic1} (English):\n", eng1[:300], "...\n")
print(f"üîπ {topic1} (French):\n", fr1[:300], "...")
```

Now let us calculate the frequency of the character n-grams. N-grams are groups of characters of size n. A unigram is a single character and a bigram is a group of two characters and so on.

Let us count the frequency of each character in a text and plot it in a histogram.

```python
# convert a tuple of characters to a string
def tuple2string(tup):
    st = ''
    for ii in tup:
        st = st + ii
    return st
# convert a tuple of tuples to a list of strings
def key2string(keys):
    return [tuple2string(i) for i in keys]
# plot the histogram
def plothistogram(ngram):
    keys = key2string(ngram.keys())
    values = list(ngram.values())
    # Sort by alphabet
    combined = zip(keys, values)
    zipped_sorted = sorted(combined, key=lambda x: x[0])
    keys, values = map(list, zip(*zipped_sorted))
    plt.figure(figsize=(10,4))
    plt.bar(keys, values)
    plt.xticks(rotation=90)
    plt.tight_layout()
```

```python
print("\nüìä Plotting UNIGRAM histograms...\n")
```

```python
unigram_eng1 = Counter(ngrams(eng1, 1))
plothistogram(unigram_eng1)
plt.title('English 1 (Giraffe) ‚Äî Unigram')
plt.show()
unigram_eng2 = Counter(ngrams(eng2, 1))
plothistogram(unigram_eng2)
plt.title('English 2 (Elephant) ‚Äî Unigram')
plt.show()
```

Another way to visualize bigrams is to use a 2-dimensional graph.

```python
def plotbihistogram(ngram, cmap='jet', normalize=False, title="Bigram Frequency Map"):
    """Plot a 2D bigram histogram with adjustable color maps and normalization."""
    freq = np.zeros((26, 26))
    total = sum(ngram.values()) if normalize else 1

    for ii in range(26):
        for jj in range(26):
            val = ngram.get((chr(ord('a') + ii), chr(ord('a') + jj)), 0)
            freq[ii, jj] = val / total if normalize else val

    plt.figure(figsize=(6, 5))
    plt.imshow(freq, cmap=cmap, origin='lower')
    plt.colorbar(label='Normalized Frequency' if normalize else 'Raw Frequency')
    plt.xticks(range(26), [chr(ord('a') + i) for i in range(26)])
    plt.yticks(range(26), [chr(ord('a') + i) for i in range(26)])
    plt.xlabel("Second Letter")
    plt.ylabel("First Letter")
    plt.title(title)
    plt.tight_layout()
    return freq
```

Let us look at the top 10 ngrams for each text.

```python
def ind2tup(ind):
    ind = int(ind)
    i = int(ind / 26)
    j = int(ind % 26)
    return (chr(ord('a') + i), chr(ord('a') + j))
def ShowTopN(bifreq, n=10, normalize=False):
    """Show top-N bigrams or n-grams with optional normalization."""
    f = bifreq.flatten()
    if normalize:
        f = f / np.sum(f)
    arg = np.argsort(-f)  # sort descending
    for ii in range(n):
        print(f'{ind2tup(arg[ii])} : {f[arg[ii]]:.5f}' if normalize else f'{ind2tup(arg[ii])} : {int(f[arg[ii]])}')
```

```python
print('\nEnglish 1 ‚Äî Top 10 Bigrams:')
ShowTopN(bieng1)
print('\nFrench 1 ‚Äî Top 10 Bigrams:')
ShowTopN(bifr1)
```

```python
print('\n English 1 ‚Äî Top 20 (Normalized):')
ShowTopN(bieng1, n=20, normalize=True)
print('\n French 1 ‚Äî Top 20 (Normalized):')
ShowTopN(bifr1, n=20, normalize=True)
```

```python
def ShowTopN(text, n=10, k=2):
    """Show top-N n-grams"""
    ngram_freq = Counter(ngrams(text, k))
    print(f"\nTop {n} {k}-grams:")
    for gram, freq in ngram_freq.most_common(n):
        print(''.join(gram), ":", freq)
    return ngram_freq
```

Let us look at another, more complicated feature. We will count the number black pixels that are surrounded on four sides by non-black pixels, or "hole pixels".

```python
def cumArray(img):
    """Cumulative sum across rows ‚Äî used for hole detection"""
    img2 = img.copy()
    for ii in range(1, img2.shape[0]):  # row-wise accumulation
        img2[ii, :] = img2[ii, :] + img2[ii-1, :]
    img2 = img2 > 0
    return img2

def getHolePixels(img):
    """Detect pixels enclosed (holes) in the digit image"""
    im1 = cumArray(img)
    im2 = np.rot90(cumArray(np.rot90(img)), 3)
    im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
    im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
    hull = im1 & im2 & im3 & im4
    hole = hull & ~(img > 0)
    return hole
```

```python
from collections import Counter
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
```

```python
def minus(a, b): return a & ~b

def cumArray(img):
    img2 = img.copy()
    for ii in range(1, img2.shape[0]):
        img2[ii, :] = img2[ii, :] + img2[ii - 1, :]
    return img2 > 0
```

```python
def get_hole_and_hull(img):
    im1 = cumArray(img)
    im2 = np.rot90(cumArray(np.rot90(img)), 3)
    im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
    im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
    hull = im1 & im2 & im3 & im4
    hole = hull & ~(img > 0)
    return hole.astype(int), hull.astype(int)
```

```python
def get_boundary(img):
    imgb = img.copy() > 0
    rshift = np.roll(imgb, 1, 1)
    lshift = np.roll(imgb, -1, 1)
    ushift = np.roll(imgb, -1, 0)
    dshift = np.roll(imgb, 1, 0)
    boundary = minus(imgb, rshift) | minus(imgb, lshift) | minus(imgb, ushift) | minus(imgb, dshift)
    return boundary.astype(int)
```

```python
def centroid(img):
    ys, xs = np.nonzero(img>0)
    if len(xs)==0:
        return (0.5, 0.5)
    return (np.mean(ys)/img.shape[0], np.mean(xs)/img.shape[1])

def second_moments(img):
    ys, xs = np.nonzero(img>0)
    if len(xs) < 2:
        return (0.0, 0.0)
    cy, cx = np.mean(ys), np.mean(xs)
    mu20 = np.mean((xs - cx)**2)
    mu02 = np.mean((ys - cy)**2)
    return (mu20, mu02)

def symmetry_features(img):
    imgb = (img>0).astype(int)
    v_sym = np.mean(imgb == np.fliplr(imgb))
    h_sym = np.mean(imgb == np.flipud(imgb))
    return h_sym, v_sym

def aspect_ratio(img):
    ys, xs = np.nonzero(img>0)
    if len(xs)==0:
        return 1.0
    h = (max(ys)-min(ys)+1)
    w = (max(xs)-min(xs)+1)
    return w / h
```

```python
def extract_features(img):
    s = np.sum(img>0)
    hole, hull = get_hole_and_hull(img)
    hole_count = hole.sum()
    hull_count = hull.sum()
    boundary_count = get_boundary(img).sum()
    cy, cx = centroid(img)
    mu20, mu02 = second_moments(img)
    h_sym, v_sym = symmetry_features(img)
    ar = aspect_ratio(img)
    return [s, hole_count, hull_count, boundary_count, cy, cx, mu20, mu02, h_sym, v_sym, ar]
```

```python
(train_X, train_y), (test_X, test_y) = mnist.load_data()
CLASSES = [0,1,2,3,4,5]  # try different sets, e.g., [0,8], or all 0-9
NUM_PER_CLASS = 400
X_list, y_list = [], []
for c in CLASSES:
    imgs = train_X[train_y == c][:NUM_PER_CLASS]
    for im in imgs:
        X_list.append(extract_features(im))
        y_list.append(c)
X = np.array(X_list)
y = np.array(y_list)
feature_names = ['sum','hole','hull','boundary','cy','cx','mu20','mu02','h_sym','v_sym','asp_ratio']
print("Feature matrix shape:", X.shape)
print("Feature names:", feature_names)
```
